```python
import pandas as pd
import numpy as np

df = pd.read_csv("data\\npf_train.csv")
df
```

Preprocessing

```python
Data = df.drop(['id','date','partlybad'], axis=1)
Data['class4'].replace(['nonevent', 'Ia', 'Ib', 'II'],[0, 1, 2, 3], inplace=True)
Data["class2"] = (Data["class4"]!= 0).astype(int)
Data
```

EDA

```python
# Class2 correlation
class2_df = Data.drop(['class4'], axis=1)
class_2_corr_matrix = class2_df.corr()
class_2_corr_matrix
```
